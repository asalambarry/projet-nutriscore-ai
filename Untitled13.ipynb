```python
import openfoodfacts
import pandas as pd

# Initialiser l'API
api = openfoodfacts.API(user_agent="MonAppNutrition/1.0")

# Champs utiles pour le modèle
def extract_features(product):
    nutriments = product.get("nutriments", {})
    return {
        "name": product.get("product_name"),
        "ingredients_text": product.get("ingredients_text"),
        "categories": product.get("categories_tags"),
        "brands": product.get("brands_tags"),
        "additives": product.get("additives_tags"),
        "nova": product.get("nova_group"),
        "energy": nutriments.get("energy_100g"),
        "sugars": nutriments.get("sugars_100g"),
        "saturated_fat": nutriments.get("saturated-fat_100g"),
        "salt": nutriments.get("salt_100g"),
        "fiber": nutriments.get("fiber_100g"),
        "proteins": nutriments.get("proteins_100g"),
        "fruits_vegetables_nuts": nutriments.get("fruits-vegetables-nuts_100g"),
        "nutrition_grade": product.get("nutrition_grades")
    }

# Liste de mots-clés à explorer (divers aliments courants)
search_terms = ["yaourt", "fromage", "jus", "céréales", "pain", "pâte", "soupe", "barre", "compote", "chocolat"]

# Collecte multi-mots-clés
all_products = []
limit_per_term = 5000
page_size = 100

for term in search_terms:
    print(f" Recherche : {term}")
    page = 1
    collected = 0
    while collected < limit_per_term:
        response = api.product.text_search(term, page=page, page_size=page_size)
        page_products = response.get("products", [])
        for p in page_products:
            data = extract_features(p)
            if data["nutrition_grade"] and data["ingredients_text"]:
                all_products.append(data)
                collected += 1
            if collected >= limit_per_term:
                break
        page += 1
        if page > response.get("page_count", 1):
            break

# Créer le DataFrame final
df = pd.DataFrame(all_products)
print(f" Total de produits récupérés : {len(df)}")
df.head()
```

```python
print("=== ANALYSE DES VALEURS MANQUANTES ===")
print("\nNombre de valeurs manquantes par colonne :")
print(df.isnull().sum())
```

```python
print("\n=== ANALYSE DES DOUBLONS ===")
print(f"\nNombre total de lignes avant traitement : {len(df)}")

# Convertir les colonnes de listes en chaînes de caractères pour la détection des doublons
df_check = df.copy()
list_columns = ['categories', 'brands', 'additives']
for col in list_columns:
    df_check[col] = df_check[col].apply(lambda x: str(x) if isinstance(x, list) else str([]))

# Maintenant nous pouvons compter les doublons
print(f"Nombre de doublons : {df_check.duplicated().sum()}")
```

# Traitement des doublons et valeurs manquantes

```python
print("\n=== TRAITEMENT DES DONNÉES ===")

# Traitement des valeurs manquantes
df_clean = df.copy()

# Remplacement des valeurs manquantes selon le type de données
df_clean['name'] = df_clean['name'].fillna('Sans nom')
df_clean['ingredients_text'] = df_clean['ingredients_text'].fillna('Non spécifié')
df_clean['categories'] = df_clean['categories'].fillna('[]').apply(lambda x: x if isinstance(x, list) else [])
df_clean['brands'] = df_clean['brands'].fillna('[]').apply(lambda x: x if isinstance(x, list) else [])
df_clean['additives'] = df_clean['additives'].fillna('[]').apply(lambda x: x if isinstance(x, list) else [])

# Remplacement des valeurs numériques manquantes par la médiane
numeric_columns = ['nova', 'energy', 'sugars', 'saturated_fat', 'salt', 'fiber', 'proteins', 'fruits_vegetables_nuts']
for col in numeric_columns:
    median_value = df_clean[col].median()
    df_clean[col] = df_clean[col].fillna(median_value)

# Suppression des doublons en utilisant df_check
duplicate_indices = df_check.duplicated()
df_clean = df_clean[~duplicate_indices]

# Affichage des résultats après traitement
print("\nNombre de valeurs manquantes après traitement :")
print(df_clean.isnull().sum())
print(f"\nNombre total de lignes après traitement : {len(df_clean)}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuration du style
plt.style.use('seaborn')
sns.set_palette("husl")

# 1. Distribution des scores nutritionnels
print("=== ANALYSE DES SCORES NUTRITIONNELS ===")

plt.figure(figsize=(12, 6))
nutrition_counts = df['nutrition_grade'].value_counts().sort_index()
sns.barplot(x=nutrition_counts.index, y=nutrition_counts.values)
plt.title('Distribution des Scores Nutritionnels (A à E)')
plt.xlabel('Score Nutritionnel')
plt.ylabel('Nombre de Produits')
for i, v in enumerate(nutrition_counts.values):
    plt.text(i, v, str(v), ha='center', va='bottom')
plt.show()

# 2. Distribution de l'énergie par score nutritionnel
plt.figure(figsize=(12, 6))
sns.boxplot(x='nutrition_grade', y='energy', data=df)
plt.title('Distribution de l\'Énergie par Score Nutritionnel')
plt.xlabel('Score Nutritionnel')
plt.ylabel('Énergie (pour 100g)')
plt.show()

# 3. Corrélation entre les nutriments
nutriments = ['energy', 'sugars', 'saturated_fat', 'salt', 'fiber', 'proteins']
plt.figure(figsize=(10, 8))
correlation = df[nutriments].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', center=0)
plt.title('Corrélation entre les Nutriments')
plt.show()

# 4. Distribution du score NOVA
plt.figure(figsize=(10, 6))
nova_counts = df['nova'].value_counts().sort_index()
sns.barplot(x=nova_counts.index, y=nova_counts.values)
plt.title('Distribution des Scores NOVA')
plt.xlabel('Score NOVA')
plt.ylabel('Nombre de Produits')
for i, v in enumerate(nova_counts.values):
    plt.text(i, v, str(int(v)), ha='center', va='bottom')
plt.show()

# 5. Top 10 des marques les plus présentes
plt.figure(figsize=(12, 6))
# Correction pour gérer les valeurs None et non-itérables
brands_list = []
for brands in df['brands']:
    if isinstance(brands, list):
        brands_list.extend(brands)
    elif isinstance(brands, str):
        brands_list.append(brands)
    # Ignorer les valeurs None et autres types non valides

brands_counts = pd.Series(brands_list).value_counts()
top_10_brands = brands_counts.head(10)
sns.barplot(x=top_10_brands.values, y=top_10_brands.index)
plt.title('Top 10 des Marques les Plus Présentes')
plt.xlabel('Nombre de Produits')
plt.show()

# 6. Distribution des nutriments clés
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
fig.suptitle('Distribution des Nutriments Clés')

nutrients = ['sugars', 'saturated_fat', 'salt', 'fiber', 'proteins', 'energy']
for i, nutrient in enumerate(nutrients):
    row = i // 3
    col = i % 3
    sns.histplot(data=df, x=nutrient, ax=axes[row, col], bins=30)
    axes[row, col].set_title(f'Distribution de {nutrient}')

plt.tight_layout()
plt.show()

# 7. Analyse des additifs
plt.figure(figsize=(12, 6))
# Correction pour gérer les valeurs None et non-itérables
additives_list = []
for additives in df['additives']:
    if isinstance(additives, list):
        additives_list.extend(additives)
    elif isinstance(additives, str):
        additives_list.append(additives)
    # Ignorer les valeurs None et autres types non valides

additives_counts = pd.Series(additives_list).value_counts()
top_10_additifs = additives_counts.head(10)
sns.barplot(x=top_10_additifs.values, y=top_10_additifs.index)
plt.title('Top 10 des Additifs les Plus Utilisés')
plt.xlabel('Nombre d\'Occurrences')
plt.show()

# Statistiques descriptives
# print("\n=== STATISTIQUES DESCRIPTIVES ===")
# print("\nStatistiques des nutriments :")
# print(df[nutriments].describe().round(2))

# print("\nRépartition des scores nutritionnels :")
# print(df['nutrition_grade'].value_counts().sort_index())

# print("\nRépartition des scores NOVA :")
# print(df['nova'].value_counts().sort_index())
```
